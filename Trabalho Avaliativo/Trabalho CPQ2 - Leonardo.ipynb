```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import control
%matplotlib inline
```

```python
# Entradas e Parâmetros:

F1 = 2.00
F2 = 3.00
C1 = 9.00
C2 = 4.00
V = 10.00
Css = 6.00

def modelo1(y,t,u):
    
    # saída (y = C) 
    C = y
           
    # Alteração na vazão com aumento de 1.38 L/min para t >= 2 min e retorna a 0 em t = 7 min
    if t < 2:  
        F1 = u
        
    if  7 > t >= 2:
        F1 = u + 1.38
        
    else:
        F1 = u
        
    # função que define o modelo
    # dydt = f(y,u,t)
        
    f = (F1*(C1-C) + F2*(C2-C))/V
    
    return f
```

```python
t0 = 0 # tempo inicial
tf = 20 # tempo final
dt = 0.01 # intervalo de tempo

# vetor de valores de tempo para integração numérica [0, 0.1, 0.2, ..., 20]
tspan = np.arange(t0,tf,dt)
n = len(tspan)

# condição inicial C(0) = Css = 6 kg/L
yss = 6
y0 = yss
```

```python
u = 2.0 # F1

y1 = odeint(modelo1, y0, tspan, args=(u,) )
```

```python
plt.figure(figsize = (8,5))
plt.title('\n \n Representação Gráfica do Modelo não-linear \n')
plt.plot(tspan,y1,'g',label='Modelo não-linear')
plt.legend()
plt.ylabel('C (kg/L)',fontsize=14)
plt.xlabel('t (min)',fontsize=14)
plt.grid(True)
plt.show()
```

```python
Kp       = K  
taup     = tau

dead_time   = 0.5  # tempo morto
d_dead_time = dead_time / dt
n_order     = 2    # da aproximação de Padé

# vamos denominar a função de transferência do tempo morto por Hdelay

num, den = control.pade(dead_time,n_order)  # determinando o num. e denom. da aprox. de Padé
Hdelay   = control.tf(num, den)             # colocando na forma de uma função de transf.

Gs = Gp * Hdelay

Kd = Kp
Gd = Kd/(taup*s+1) * Hdelay
```

### Função de transferência em malha fechada

A função de transferência em malha fechada é determinada de acordo com a Equação abaixo:

\begin{equation}
Y(s) = \dfrac{G_c(s) G(s)}{1 + G_c(s) G(s)} Y_{SP}(s) + \dfrac{G_d(s)}{1 + G_c(s) G(s)} L(s)
\end{equation}

Para simulação do problema servo, utilizamos:

\begin{equation}
\dfrac{Y(s)}{ Y_{SP}(s)} = \dfrac{G_c(s) G(s)}{1 + G_c(s) G(s)} 
\end{equation}

Considerando agora o problema regulador, temos:

\begin{equation}
\dfrac{Y(s)}{ L(s)} = \dfrac{G_d(s)}{1 + G_c(s) G(s)} 
\end{equation}

Para investigarmos o comportamento em malha fechada, simulamos os efeitos separados. 

### Simulando o problema servo:

```python
def pid_servo(Kc, tauI, tauD, sint):

    t = np.arange(t0,tf,dt) 
    
    # inicializando e definindo a trajetória de referência
    SP = np.zeros(n) 
    SP[1:n] = 1.0     
    
    # cálculo de Ki = Kc/tauI e Kd = tauD*Kc
    Ki = Kc/tauI
    Kd = Kc*tauD

    # Determinando as ações de controle de forma isolada
    P = Kc
    I = Ki/s 
    D = Kd*s

    # Somando os termos proporcional, integral e derivativo
    Gc = P+I+D 
        
    Gmf = Gs*Gc/(1+Gc*Gs)
    
    t, y = control.forced_response (Gmf, t, SP)
    
    plt.plot(t, y, label=sint, scalex = 6, scaley = 6)
    plt.plot(t, SP, 'k--')
    plt.ylim(-0.2, 2.0)
    plt.title('\n Problema servo (alteração degrau no setpoint) \n')
    plt.ylabel('$y(t)$,$y_{sp}(t)$')
    plt.xlabel('tempo (min)')
    plt.grid('on',linestyle=':')
    plt.legend()
```

```python
# cálculo do PI servo para diferentes os parâmetrs de sintonia
plt.figure(figsize = (9,5))
pid_servo (Kc=1.27, tauI=5.93, tauD=0, sint= "a) Ziegler e Nichols (1º Método)")
pid_servo (Kc=1.80, tauI=5.93, tauD=0, sint= "b) ITAE Regulador")
pid_servo (Kc=2.97, tauI=2.75, tauD=0, sint= "c) ITAE Servo")
pid_servo (Kc=1.83, tauI=5.93, tauD=0, sint= "d) TauC = ")
```

### Simulando o problema regulador

Utilizando agora a função $G_d(s)$ e considerando que a perturbação $L(s)$ entra no sistema:

```python
def pid_regulador(Kc,tauI,tauD,sint):

    t = np.arange(t0,tf,dt) # create time vector   
    n = len(t)
    
    L  = np.zeros(n) 
    SP = np.zeros(n) 
    
    SP[0:n] = 0.0              # define setpoint
    L[538: ] = 1.0     # define setpoint
    
    Ki = Kc/tauI
    Kd = Kc*tauD

    P = Kc
    I = Ki/s 
    D = Kd*s

    Gc = P+I+D 
        
    Gmf = Gd/(1+Gc*Gs)
    t,y = control.forced_response(Gmf,t,L)
    
    plt.subplot(2, 1, 1)
    plt.plot(t, y, label = sint)
    plt.plot(t, SP, 'k--')
    plt.ylim(-0.5, 0.5)
    plt.title('\n Problema regulador (alteração degrau na carga/perturbação)\n')
    plt.ylabel('$y_{(t)}$, $y_{sp(t)}$')
    plt.xlabel('tempo (min)')
    plt.grid('on',linestyle=':')
    plt.legend()
```
